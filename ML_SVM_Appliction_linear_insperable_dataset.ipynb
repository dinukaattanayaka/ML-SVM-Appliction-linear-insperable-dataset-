```python
import pandas as pd

#dataset=pd.read_csv(r'C:\Users\Migara Liyanage\Desktop\ML 5 Day Workshop 4\MLW-Day-01-main\iris.csv')
dataset=pd.read_csv('/content/drive/MyDrive/machine learning/DAY 2/linear-insperable-dataset.csv').values

print(dataset,dataset.shape)
```

```python
data = dataset[:,0:2] #all raws and 0 & 1 ccolumns
target = dataset[:,2]
print(data,target)
```

```python
print(data.shape)
```

```python
from matplotlib import pyplot as plt

x1 = data[:,0]
x2 = data[:,1]

for i in range(len(data)):

  if target[i] == 0:
    plt.scatter(x1[i],x2[i],color='red')
  elif target[i] == 1:
    plt.scatter(x1[i],x2[i],color='green')

plt.xlabel('feature 1(x1)')
plt.ylabel('feature 2(x2)')
plt.show()
```

```python
from sklearn.model_selection import train_test_split #sklearn library eke model_selection sub module eke train_test_split function

train_data,test_data,train_target,test_target=train_test_split(data,target,test_size=0.2)
```

```python
from sklearn.svm import SVC #support vector clssifier eka gnnw

model=SVC(kernel='rbf')
#model=SVC(kernel='poly' degree=2)        #polynomial,RBF,Gausian Basis,Sigmoid,Tanh
model.fit(train_data,train_target)
```

```python
#evaluating/testing
predicted_target=model.predict(test_data)
```

```python
from sklearn.metrics import accuracy_score

acc=accuracy_score(test_target,predicted_target)
print('accuracy:',acc)
```
